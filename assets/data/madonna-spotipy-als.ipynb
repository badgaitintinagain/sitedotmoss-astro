# Step 1: Data Cleaning

```python
# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 1: DATA CLEANING & PREPROCESSING (IMPROVED)
# ═══════════════════════════════════════════════════════════════════════════════

print("DATA CLEANING & PREPROCESSING")
print("="*80)

# Original count
original_count = len(madonna_tracks)
print(f"Original tracks: {original_count}")

# Define audio features
audio_features = ['danceability', 'energy', 'valence', 'tempo', 'acousticness', 
                  'loudness', 'mode', 'key', 'speechiness', 'instrumentalness', 'liveness']

# Step 1: Remove NULL values (before anything else)
madonna_cleaned = madonna_tracks.dropna(subset=audio_features[:5]).copy()
print(f"After removing NaN: {len(madonna_cleaned)} tracks")

# Step 2: Remove non-studio versions
exclude_keywords = r'live|remaster|mix|edit|version|dub|remix|acoustic|instrumental|cover|tribute|mashup'
madonna_cleaned = madonna_cleaned[~madonna_cleaned['name'].str.contains(exclude_keywords, case=False, na=False)]
print(f"After removing alternate versions: {len(madonna_cleaned)} tracks")

# Step 3: Remove exact duplicates (keep earliest release)
madonna_cleaned = madonna_cleaned.sort_values('release_date').drop_duplicates(subset=['name'], keep='first')
print(f"After removing duplicates: {len(madonna_cleaned)} tracks")

# Step 4: Extract year and decade
madonna_cleaned['release_year'] = pd.to_datetime(madonna_cleaned['release_date'], errors='coerce').dt.year
madonna_cleaned['decade'] = (madonna_cleaned['release_year'] // 10 * 10).astype(int)

# Remove any rows with invalid years
madonna_cleaned = madonna_cleaned[(madonna_cleaned['release_year'] >= 1980) & (madonna_cleaned['release_year'] <= 2025)]
print(f"After filtering date range: {len(madonna_cleaned)} tracks")

print(f"\nFINAL: {len(madonna_cleaned)} tracks ready for analysis")
print(f"Year range: {madonna_cleaned['release_year'].min()} - {madonna_cleaned['release_year'].max()}")
print(f"Tracks by decade:\n{madonna_cleaned['decade'].value_counts().sort_index()}")
```

# -> K-Mean Clustering

```python
# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 3: K-MEANS CLUSTERING WITH OPTIMIZATION
# ═══════════════════════════════════════════════════════════════════════════════

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

print("\nK-MEANS CLUSTERING ANALYSIS")
print("="*80)

# Step 1: Select and scale features
clustering_features = ['danceability', 'energy', 'valence', 'acousticness', 'loudness']
X_cluster = madonna_cleaned[clustering_features].copy()

# Scale features to same range
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

print(f"Features used for clustering: {clustering_features}")
print(f"Dataset shape: {X_scaled.shape}")

# Step 2: Elbow Method to find optimal clusters
print("\nFinding optimal number of clusters (Elbow Method)...")
inertias = []
silhouette_scores = []
davies_bouldin_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_temp = kmeans_temp.fit_predict(X_scaled)
    inertias.append(kmeans_temp.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, labels_temp))
    davies_bouldin_scores.append(davies_bouldin_score(X_scaled, labels_temp))

# Plot Elbow curve
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# Elbow plot
axes[0].plot(K_range, inertias, marker='o', linewidth=2, markersize=8, color='#FF1493')
axes[0].set_xlabel('Number of Clusters (K)')
axes[0].set_ylabel('Inertia (Within-cluster sum of squares)')
axes[0].set_title('Elbow Method for Optimal K', fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Elbow point', xy=(4, inertias[2]), xytext=(5, inertias[2]+10),
                arrowprops=dict(arrowstyle='->', color='red'), fontsize=10, color='red')

# Silhouette score plot
axes[1].plot(K_range, silhouette_scores, marker='s', linewidth=2, markersize=8, color='#1DB954')
axes[1].set_xlabel('Number of Clusters (K)')
axes[1].set_ylabel('Silhouette Score')
axes[1].set_title('Silhouette Score (Higher is Better)', fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].axhline(y=max(silhouette_scores), color='green', linestyle='--', alpha=0.5, label='Best')
axes[1].legend()

# Davies-Bouldin score plot
axes[2].plot(K_range, davies_bouldin_scores, marker='^', linewidth=2, markersize=8, color='#1ed760')
axes[2].set_xlabel('Number of Clusters (K)')
axes[2].set_ylabel('Davies-Bouldin Index')
axes[2].set_title('Davies-Bouldin Index (Lower is Better)', fontweight='bold')
axes[2].grid(True, alpha=0.3)
axes[2].axhline(y=min(davies_bouldin_scores), color='green', linestyle='--', alpha=0.5, label='Best')
axes[2].legend()

plt.tight_layout()
plt.show()

optimal_k = 4
print(f"\nOptimal K selected: {optimal_k}")
print(f"   Silhouette Score: {silhouette_scores[optimal_k-2]:.4f}")
```

```python
# Step 3: Final K-Means with K=4
print(f"\nTraining K-Means with {optimal_k} clusters...")

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10, max_iter=300)
madonna_cleaned['cluster'] = kmeans.fit_predict(X_scaled)

print(f"Clustering complete!")
print(f"\nCluster Distribution:")
print(madonna_cleaned['cluster'].value_counts().sort_index())

# Step 4: Analyze cluster characteristics
print("\nCLUSTER PROFILES")
print("-"*80)

cluster_profiles = madonna_cleaned.groupby('cluster')[clustering_features + ['release_year']].agg(['mean', 'std', 'min', 'max'])

for cluster_id in range(optimal_k):
    cluster_data = madonna_cleaned[madonna_cleaned['cluster'] == cluster_id]
    
    print(f"\n🎵 CLUSTER {cluster_id}")
    print(f"   Size: {len(cluster_data)} tracks")
    print(f"   Year range: {cluster_data['release_year'].min():.0f} - {cluster_data['release_year'].max():.0f}")
    print(f"   Danceability: {cluster_data['danceability'].mean():.3f}")
    print(f"   Energy: {cluster_data['energy'].mean():.3f}")
    print(f"   Valence: {cluster_data['valence'].mean():.3f}")
    print(f"   Acousticness: {cluster_data['acousticness'].mean():.3f}")
    print(f"   Loudness: {cluster_data['loudness'].mean():.2f} dB")
    print(f"   Popularity: {cluster_data['popularity'].mean():.1f}")
    
    # Show representative tracks
    top_tracks = cluster_data.nlargest(3, 'popularity')[['name', 'release_year', 'popularity']]
    print(f"   Top tracks:")
    for _, track in top_tracks.iterrows():
        print(f"      • {track['name']} ({int(track['release_year'])}) - Pop: {track['popularity']:.0f}")
```

```python
# Step 5: Advanced K-Means Visualization
print("\nK-MEANS VISUALIZATION")
print("="*80)

fig, axes = plt.subplots(2, 2, figsize=(16, 13))

# Plot 1: Danceability vs Energy
ax1 = axes[0, 0]
for cluster in range(optimal_k):
    cluster_data = madonna_cleaned[madonna_cleaned['cluster'] == cluster]
    ax1.scatter(cluster_data['danceability'], cluster_data['energy'], 
               label=f'Cluster {cluster}', s=120, alpha=0.7, edgecolors='white', linewidth=0.5)

# Plot cluster centers
centers_2d = scaler.inverse_transform(kmeans.cluster_centers_)
ax1.scatter(centers_2d[:, 0], centers_2d[:, 1], marker='*', s=1000, c='red', 
           edgecolors='black', linewidth=2, label='Centroids', zorder=5)

ax1.set_xlabel('Danceability', fontweight='bold')
ax1.set_ylabel('Energy', fontweight='bold')
ax1.set_title('K-Means Clustering: Danceability vs Energy', fontweight='bold', fontsize=12)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Valence vs Acousticness
ax2 = axes[0, 1]
for cluster in range(optimal_k):
    cluster_data = madonna_cleaned[madonna_cleaned['cluster'] == cluster]
    ax2.scatter(cluster_data['valence'], cluster_data['acousticness'], 
               label=f'Cluster {cluster}', s=120, alpha=0.7, edgecolors='white', linewidth=0.5)

ax2.set_xlabel('Valence (Positivity)', fontweight='bold')
ax2.set_ylabel('Acousticness', fontweight='bold')
ax2.set_title('K-Means Clustering: Valence vs Acousticness', fontweight='bold', fontsize=12)
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Cluster timeline
ax3 = axes[1, 0]
for cluster in range(optimal_k):
    cluster_data = madonna_cleaned[madonna_cleaned['cluster'] == cluster]
    year_cluster = cluster_data.groupby('release_year').size()
    ax3.plot(year_cluster.index, year_cluster.values, marker='o', label=f'Cluster {cluster}', linewidth=2, markersize=6)

ax3.set_xlabel('Release Year', fontweight='bold')
ax3.set_ylabel('Number of Tracks', fontweight='bold')
ax3.set_title('Cluster Distribution Over Time', fontweight='bold', fontsize=12)
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Cluster characteristics (radar-like)
ax4 = axes[1, 1]
cluster_means = madonna_cleaned.groupby('cluster')[clustering_features].mean()
x = np.arange(len(clustering_features))
width = 0.2

for cluster in range(optimal_k):
    ax4.bar(x + cluster*width, cluster_means.loc[cluster].values, 
           width, label=f'Cluster {cluster}', alpha=0.8)

ax4.set_xlabel('Features', fontweight='bold')
ax4.set_ylabel('Average Value', fontweight='bold')
ax4.set_title('Cluster Feature Profiles Comparison', fontweight='bold', fontsize=12)
ax4.set_xticks(x + width * 1.5)
ax4.set_xticklabels(clustering_features, rotation=45, ha='right')
ax4.legend()
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

# -> HDBSCAN

```python
# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 4: HDBSCAN (DENSITY-BASED CLUSTERING)
# ═══════════════════════════════════════════════════════════════════════════════

from sklearn.cluster import HDBSCAN

print("\nHDBSCAN CLUSTERING ANALYSIS (Density-Based)")
print("="*80)

# Train HDBSCAN
hdbscan_model = HDBSCAN(min_cluster_size=5, min_samples=3)
madonna_cleaned['hdbscan_cluster'] = hdbscan_model.fit_predict(X_scaled)

n_clusters_hdbscan = len(set(madonna_cleaned['hdbscan_cluster'])) - (1 if -1 in madonna_cleaned['hdbscan_cluster'] else 0)
n_noise = list(madonna_cleaned['hdbscan_cluster']).count(-1)

print(f"HDBSCAN complete!")
print(f"   Clusters found: {n_clusters_hdbscan}")
print(f"   Noise points (outliers): {n_noise} ({n_noise/len(madonna_cleaned)*100:.1f}%)")
print(f"\nCluster Distribution:")
print(madonna_cleaned['hdbscan_cluster'].value_counts().sort_index())

# Visualization comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# K-Means
ax1 = axes[0]
scatter1 = ax1.scatter(madonna_cleaned['danceability'], madonna_cleaned['energy'], 
                       c=madonna_cleaned['cluster'], cmap='viridis', s=100, alpha=0.7, edgecolors='white', linewidth=0.5)
ax1.set_xlabel('Danceability', fontweight='bold')
ax1.set_ylabel('Energy', fontweight='bold')
ax1.set_title('K-Means Clustering (4 clusters)', fontweight='bold', fontsize=12)
plt.colorbar(scatter1, ax=ax1, label='Cluster')

# HDBSCAN
ax2 = axes[1]
# Create color map for HDBSCAN (noise points in gray)
colors = np.where(madonna_cleaned['hdbscan_cluster'] == -1, 0.5, madonna_cleaned['hdbscan_cluster'])
scatter2 = ax2.scatter(madonna_cleaned['danceability'], madonna_cleaned['energy'], 
                       c=colors, cmap='tab10', s=100, alpha=0.7, edgecolors='white', linewidth=0.5)
ax2.set_xlabel('Danceability', fontweight='bold')
ax2.set_ylabel('Energy', fontweight='bold')
ax2.set_title(f'HDBSCAN Clustering ({n_clusters_hdbscan} clusters + noise)', fontweight='bold', fontsize=12)
plt.colorbar(scatter2, ax=ax2, label='Cluster (-1 = Noise)')

plt.tight_layout()
plt.show()

print("\nHDBSCAN visualization complete!")
```

# -> Dimensity Reduction

```python
# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 5: DIMENSIONALITY REDUCTION (PCA & t-SNE)
# ═══════════════════════════════════════════════════════════════════════════════

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import plotly.express as px
import plotly.graph_objects as go

print("\nDIMENSIONALITY REDUCTION (PCA & t-SNE)")
print("="*80)

# Step 1: PCA Analysis
print("\nPerforming PCA...")
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance
cumsum_var = np.cumsum(pca.explained_variance_ratio_)

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Variance explained
axes[0].bar(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, alpha=0.7, label='Individual')
axes[0].plot(range(1, len(cumsum_var)+1), cumsum_var, 'ro-', linewidth=2, markersize=6, label='Cumulative')
axes[0].axhline(y=0.95, color='green', linestyle='--', linewidth=2, label='95% threshold')
axes[0].set_xlabel('Principal Component', fontweight='bold')
axes[0].set_ylabel('Explained Variance Ratio', fontweight='bold')
axes[0].set_title('PCA Explained Variance', fontweight='bold', fontsize=12)
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# PC1 vs PC2
scatter = axes[1].scatter(X_pca[:, 0], X_pca[:, 1], c=madonna_cleaned['cluster'], 
                         cmap='viridis', s=100, alpha=0.7, edgecolors='white', linewidth=0.5)
axes[1].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)', fontweight='bold')
axes[1].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)', fontweight='bold')
axes[1].set_title('PCA: First 2 Principal Components', fontweight='bold', fontsize=12)
plt.colorbar(scatter, ax=axes[1], label='Cluster')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"PCA complete!")
print(f"   First 2 PCs explain: {cumsum_var[1]*100:.1f}% of variance")
print(f"   First 3 PCs explain: {cumsum_var[2]*100:.1f}% of variance")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# ADVANCED FEATURE ENGINEERING & ANALYSIS
# ═══════════════════════════════════════════════════════════════════════════════

print("🎨 ADVANCED FEATURE ENGINEERING")
print("="*80)

# Create sophisticated features from existing data
madonna_cleaned['loudness_normalized'] = (madonna_cleaned['loudness'] - madonna_cleaned['loudness'].min()) / (madonna_cleaned['loudness'].max() - madonna_cleaned['loudness'].min())
madonna_cleaned['tempo_category'] = pd.cut(madonna_cleaned['tempo'], bins=[0, 90, 120, 150, 300], labels=['Slow', 'Moderate', 'Fast', 'Very Fast'])
madonna_cleaned['key_name'] = madonna_cleaned['key'].map({
    0: 'C', 1: 'C#', 2: 'D', 3: 'D#', 4: 'E', 5: 'F', 6: 'F#', 7: 'G', 8: 'G#', 9: 'A', 10: 'A#', 11: 'B'
})
madonna_cleaned['mode_name'] = madonna_cleaned['mode'].map({0: 'Minor', 1: 'Major'})
madonna_cleaned['duration_min'] = madonna_cleaned['duration_ms'] / 60000

# Create "Sonic Energy Index" (combined metric)
madonna_cleaned['sonic_energy_index'] = (
    madonna_cleaned['energy'] * 0.35 + 
    madonna_cleaned['loudness_normalized'] * 0.25 + 
    madonna_cleaned['danceability'] * 0.25 +
    madonna_cleaned['valence'] * 0.15
)

# Create "Emotional Depth Index"
madonna_cleaned['emotional_depth'] = (
    madonna_cleaned['valence'] * -0.3 +  # Lower valence = deeper
    madonna_cleaned['acousticness'] * 0.4 +
    madonna_cleaned['instrumentalness'] * 0.2 +
    (1 - madonna_cleaned['energy']) * 0.1
)

# Track "Sonic DNA" - fingerprint combining all features
madonna_cleaned['sonic_complexity'] = (
    madonna_cleaned['instrumentalness'] + 
    madonna_cleaned['liveness'] + 
    madonna_cleaned['speechiness']
) / 3

print("✅ Advanced features created:")
print(f"   • Sonic Energy Index: Combined energy metric")
print(f"   • Emotional Depth: Emotional intensity")
print(f"   • Sonic Complexity: Instrumentation diversity")
print(f"   • Duration categories: Fast/Moderate/Slow")
print(f"   • Key distribution: Musical key analysis")
print(f"   • Mode distribution: Major/Minor analysis")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# ADVANCED DIVA COMPARISON - MULTI-ARTIST ANALYSIS
# ═══════════════════════════════════════════════════════════════════════════════

print("\n👑 ADVANCED DIVA COMPARISON")
print("="*80)

# Define divas + more context
true_divas = ['Madonna', 'Britney Spears', 'Mariah Carey', 'Kylie Minogue', 'Whitney Houston',
              'Lady Gaga', 'Adele', 'Taylor Swift', 'Rihanna', 'Christina Aguilera']

# Get tracks
diva_all_tracks = tracks[tracks['artists'].str.contains('|'.join(true_divas), na=False, case=False)].copy()

# Calculate rich profiles
diva_features = ['danceability', 'energy', 'valence', 'acousticness', 'loudness', 'tempo', 'popularity']
diva_profiles_full = diva_all_tracks.groupby('artists')[diva_features].agg(['mean', 'std', 'min', 'max']).round(3)

# Clean and filter
diva_clean_profiles = diva_all_tracks.groupby('artists')[diva_features].mean().reset_index()
diva_clean_profiles['artists'] = diva_clean_profiles['artists'].str.replace(r'["\'\[\]]', '', regex=True)
diva_clean_profiles = diva_clean_profiles[diva_clean_profiles['artists'].isin(true_divas)].head(10)

print("\n📊 DIVA AUDIO PROFILES (Complete Statistics)")
print("-"*80)

for _, diva_row in diva_clean_profiles.iterrows():
    diva_name = diva_row['artists']
    print(f"\n🎤 {diva_name}")
    print(f"   Danceability: {diva_row['danceability']:.3f}")
    print(f"   Energy: {diva_row['energy']:.3f}")
    print(f"   Valence: {diva_row['valence']:.3f}")
    print(f"   Acousticness: {diva_row['acousticness']:.3f}")
    print(f"   Loudness: {diva_row['loudness']:.2f} dB")
    print(f"   Tempo: {diva_row['tempo']:.0f} BPM")
    print(f"   Avg Popularity: {diva_row['popularity']:.1f}/100")

# Create comprehensive comparison viz
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

comparison_features = ['danceability', 'energy', 'valence', 'acousticness', 'popularity']

for idx, feature in enumerate(comparison_features):
    ax = axes[idx // 3, idx % 3]
    
    sorted_divas = diva_clean_profiles.sort_values(feature, ascending=False).head(8)
    colors_bar = plt.cm.viridis(np.linspace(0, 1, len(sorted_divas)))
    
    bars = ax.barh(sorted_divas['artists'], sorted_divas[feature], color=colors_bar, edgecolor='white', linewidth=1)
    ax.set_xlabel(feature.capitalize(), fontweight='bold')
    ax.set_title(f'Divas Ranked by {feature.capitalize()}', fontweight='bold', fontsize=11)
    ax.grid(True, alpha=0.3, axis='x')
    
    # Add value labels
    for i, (bar, value) in enumerate(zip(bars, sorted_divas[feature])):
        ax.text(value, bar.get_y() + bar.get_height()/2, f'{value:.3f}', 
               ha='left', va='center', fontsize=9, fontweight='bold')

# Last subplot: Madonna positioning
ax = axes[1, 2]
ax.axis('off')

madonna_stats = diva_clean_profiles[diva_clean_profiles['artists'] == 'Madonna'].iloc[0]
stats_text = f"""
👑 MADONNA POSITIONING

Compared to Top 10 Pop Divas:

Danceability: {madonna_stats['danceability']:.3f}
  Rank: {(diva_clean_profiles['danceability'] > madonna_stats['danceability']).sum() + 1}/10

Energy: {madonna_stats['energy']:.3f}
  Rank: {(diva_clean_profiles['energy'] > madonna_stats['energy']).sum() + 1}/10

Valence: {madonna_stats['valence']:.3f}
  Rank: {(diva_clean_profiles['valence'] > madonna_stats['valence']).sum() + 1}/10

Popularity: {madonna_stats['popularity']:.1f}
  Rank: {(diva_clean_profiles['popularity'] > madonna_stats['popularity']).sum() + 1}/10

🎯 KEY INSIGHT:
Madonna ranks in top positions for:
- Danceability (essential to her brand)
- Energy (high-intensity productions)

She differs by being less acoustic,
more electronic & produced.
"""

ax.text(0.05, 0.95, stats_text, transform=ax.transAxes, fontsize=10,
       verticalalignment='top', fontfamily='monospace',
       bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

plt.suptitle('👑 Advanced Diva Comparison Dashboard', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n✅ Diva comparison complete!")
```
